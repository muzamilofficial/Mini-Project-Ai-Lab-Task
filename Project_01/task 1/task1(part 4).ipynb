```python
print("Welcome To Hangman Game")
import random

f=open("words.txt","r")
data=f.read()
words=data.split()
word=random.choice(words)
word=word.upper()
total_chances=7
guss_word="_"*len(word)
print(guss_word)

while total_chances!=0:
    # print(guss_word)
    letter =input("enter your letter=").upper()
    if letter in word:
        for index in range(len(word)):
            if word[index]==letter:
             guss_word=guss_word[:index]+letter+guss_word[index+1:]
        print(guss_word)
        
        if guss_word==word:
            print("congrats you win")
            print("your correct word is ",word)
            break
    else:
        total_chances-=1
        print("your letter is not correct")
        print("your remaining chances ",total_chances)
        if total_chances==6:
            print("......|......")
        elif total_chances==5:
            print(".....|.....")
            print(".....O.....")
        elif total_chances==4:
            print("......|.......")
            print("......O.......")
            print("......|.......")
        elif total_chances==3:
            print("......|.......")
            print("......O.......")
            print("...../|.......")
        elif total_chances==2:
            print("......|.......")
            print("......O.......")
            print("...../|\\......")
        elif total_chances==1:
            print("......|.......")
            print("......O.......")
            print("...../|\\......")
            print("...../.........")
        elif total_chances==0:
            print("......|.......")
            print("......O.......")
            print("...../|\......")
            print("...../.\\......")
else:
    print("you lose")
    print("try again")
```
